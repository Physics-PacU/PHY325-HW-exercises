```python
# To begin I will be making example 3.1 in Newman 
#an object oriented plot

# So to start I have to generate some data
pi = math.pi
#I specify which library I am using to get pi
theta = np.linspace(0,2*pi,100)
#Then I use linspace to make values in a reasonable domain
sin_theta = np.sin(theta)
#Next I use a numpy function for my y-values

fig, ax = plt.subplots()
#To make an object oriented plot I define my axes and figure
ax.plot(theta,sin_theta,'-',c='m')

#Then I individually set the labels as attributes of the axes object
ax.set_xlabel("theta")
ax.set_ylabel("sin(theta)")
ax.set_title("Newman 3.1 example, Graph of the sine function")
```

```python
#Here is my object oriented version of the above code


wavelength = 5.0
k = 2*pi/wavelength
xi0 = 1.0
separation = 20.0
side = 100.0
points = 500
spacing = side/points  
# for the code of the wave interactions
#I am not really changing anything since it is sound. 

x1 = side/2 + separation/2
y1 = side/2
x2 = side/2 - separation/2
y2 = side/2

xi = empty([points,points],float)
# Now we have the array of the heights
#Same thing still as Newman's example


for i in range(points):
    y = spacing*i
    for j in range(points):
        x = spacing*j
        r1 = sqrt((x-x1)**2+(y-y1)**2)
        r2 = sqrt((x-x2)**2+(y-y2)**2)
        xi[i,j] = xi0*sin(k*r1) + xi0*sin(k*r2)
        #This sets each particular point in the array
        
fig, ax = plt.subplots(figsize=(8,8))
#In creating the fig and ax objects, 
#I made the attribute figsize larger so I could better read the titles/labels
ax.set_xlabel("horizontal distance (cm)")
ax.set_ylabel("vertical distance (cm)")
ax.set_title("Interference pattern of two waves separated by {} cm".format(separation))
#And here is the formatting method we learned on Monday

ax.imshow(xi,'spring',origin="lower",extent=[0,side,0,side])
```

```python
#Now for Newman excercise 3.3
#The STM data

STM_data= np.loadtxt("stm.txt",float)

fig, ax = plt.subplots()
ax.set_title("Scanning Tunneling Microscope Data")

close = mpatches.Patch(color='yellow', label='closer surface')
far = mpatches.Patch(color='purple', label='farther surface')
#The line below adds a legend to the plot
# I made two quantitative handles to make the visual easier to interpret
#Then I set the legend attribute of the ax object
#using the handles
#Also I used the bb_to_anchor parameter to keep the legend off of the figure
ax.legend(handles=[close,far],bbox_to_anchor=(1.05, 1),loc='upper left')

ax.imshow(STM_data,'plasma')
```
